### urdu2roman()

```python
def urdu2roman(stringg):
    #dictionary that contains all roman urdu equivalents of each urdu character
    urdu_roman = {
        'ا': 'a',
        'آ': 'aa',
        'ب': 'b',
        'پ': 'p',
        'ت': 't',
        'ٹ': 'tt',
        'ث': 's',
        'ج': 'j',
        'چ': 'ch',
        'ح': 'h',
        'خ': 'kh',
        'د': 'd',
        'ڈ': 'dd',
        'ذ': 'z',
        'ر': 'r',
        'ڑ': 'rr',
        'ز': 'z',
        'ژ': 'zh',
        'س': 's',
        'ش': 'sh',
        'ص': 'ss',
        'ض': 'z',
        'ط': 't',
        'ظ': 'z',
        'ع': 'aa',
        'غ': 'gh',
        'ف': 'f',
        'ق': 'q',
        'ک': 'k',
        'گ': 'g',
        'ل': 'l',
        'م': 'm',
        'ن': 'n',
        'ں': 'n',
        'و': 'o',
        'ہ': 'h',
        'ی': 'y',
        'ے': 'ey'
    }
    
    #make a translation table from the dictionary urdu_roman that maps roman urdu char to their corresponding urdu char
    translated_tokens = []
    
    tokens = stringg[0].split()
    for words in tokens:
        
        roman = ""
        
        for letter in words:
            
            if letter in urdu_roman:
                
                roman = roman + urdu_roman[letter]
                
        
        translated_tokens.append(roman)
        
                
        
#     translation_table = str.maketrans(urdu_roman)
 
#     #convert using the translation table
#     translated_tokens = []
#     for i in stringg:
#         translated_token = i.translate(translation_table)
#         translated_tokens.append(translated_token)
 
    text = ' '.join(translated_tokens) #array to string
    
    #returns translated text as string
    return text
```

```python
f = file2[4]
print("\nOriginal text:\n",f)
arr = urdu2roman(f)
print("\nFunction Returns(string):\n",arr)
```

```python
def evaluate(actual, translated):
    # Create empty dictionaries to store missing, extra, and correct words
    missing = 0
    extra = 0
    correct = 0
    
    if len(actual) < len(translated):
        L = len(actual)     
    else:
        L = len(translated)

    # Iterate through each word in actual and translated
    for i in range(L):
        # Remove any empty strings
        if actual[i] == '' or translated[i] == '':
            continue
        
        # Compare the lengths of the actual and translated words
        if len(actual[i]) > len(translated[i]):
            # number of missing letters
            missing +=1 
        
        elif len(actual[i]) < len(translated[i]):
            # number of extra letters added in words
            extra +=1 
        
        else:
            # correct words
            correct+=1
    
    # Calculate accuracy as the percentage of correct words out of total words
    accuracy = (correct) / len(actual) * 100
    
    
    # Return dictionaries for missing, extra, and correct words, and accuracy percentage
    return missing, extra, correct, accuracy

t = urdu2roman(f)
print("Tranlated:\n",urdu2roman(f))
print("Actual: ",file1[4])
Miss,Extra,Correct,accuracy = evaluate(file1[4][0].split(' '),t.split())
print("Missing: ",(Miss))
print("Extra: ",(Extra))
print("correct: ",(Correct))
print("Final Evaluation: ", accuracy)
```
